### Case - Cancelamento de Clientes

Você foi contratado por uma empresa com mais de 800 mil clientes para um projeto de Dados. Recentemente a empresa percebeu que da sua base total de clientes, a maioria são clientes inativos, ou seja, que já cancelaram o serviço.

Precisando melhorar seus resultados ela quer conseguir entender os principais motivos desses cancelamentos e quais as ações mais eficientes para reduzir esse número.

Base de dados e arquivos: https://www.kaggle.com/datasets/lucasoliveiramoro/cancelamentos?form=MG0AV3

```python
# Passo a passo
# instale as dependências necessarias
# !pip install pandas numpy openpyxl nbformat ipykernel plotly

# passo 1: Importar a base de dados
import pandas as pd
tabela = pd.read_csv("cancelamentos_sample.csv")

# passo 2: Visualizar a base de dados (2 objetivos: entender as info e encontrar problemas)
#display(tabela) vizualize a tabela e remova colunas desnecessárias
# Tirar informações inúteis
tabela = tabela.drop(columns=["CustomerID"]) # Remover a coluna CustomerID, pois não é necessária para a análise
display(tabela)

```

```python
# passo 3: Resolver os problemas da base de dados
# Verificar se existem valores nulos
# Verificar se existem informações formato errado
display(tabela.info())
tabela = tabela.dropna()
display(tabela.info())
```

```python
# passo 4: Análise Inicial 
# quantos clientes cancelaram
display(tabela["cancelou"].value_counts())

# qual o % de clientes que cancelaram
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

```python
# passo 5: Análise a causa dos cancelamentos dos clientes
# como cada coluna impacta no cancelamento
# criando graficos para cada coluna
import plotly.express as px
# use toda a tabela para a ánalise
"""
for coluna in tabela.columns:
    grafico = px.histogram(tabela, x=coluna, color="cancelou", text_auto=True, title=coluna)
    grafico.show()
"""

# aqui usei apenas algumas colunas para não poluir a tela
colunas_para_graficos = ["ligacoes_callcenter", "dias_atraso", "assinatura", "duracao_contrato", "cancelou"]
for coluna in colunas_para_graficos:
    grafico = px.histogram(tabela, x=coluna, color="cancelou", text_auto=True, title=coluna)
    grafico.show()
```

### Análise dos graficos

Analisando os gráficos, percebe-se três grandes problemas. Todos os clientes que ligam mais de quatro vezes para o callcenter acabam cancelando a assinatura. Clientes do contrato mensal todos cancelam. E clientes que atrasam mais de 20 dias também cancelaram.

Vamos então resolver esses 3 problemas: criando um processo para resolver o problema do cliente em no máximo 3 ligações, gerando um plano de desconto para assinaturas anuais e trimestrais e gerando uma política para resolver atrasos em até 10 dias.

```python
# clientes que ligam mais do que 4 vezes para o call center, cancelam
    # criar um processo para resolver o problema do cliente em no máximo 3 ligações
tabela = tabela[tabela["ligacoes_callcenter"] <= 4]

# clientes que atrasaram mais de 20 dias, cancelaram
    # política de resolver atrasos em até 10 dias (equipe financeira)
tabela = tabela[tabela["dias_atraso"] <= 20]

# clientes do contrato mensal TODOS cancelam
    # política para desconto para assinatura anual
tabela = tabela[tabela["duracao_contrato"] != "Monthly"]

# numero de clientes após as 3 condições
display(tabela["cancelou"].value_counts())
# porcentagem de clientes após as 3 condições
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

### 5. Resultados

Após aplicar as ações propostas, é possível observar uma redução significativa no número e no percentual de cancelamentos, que eram de 56,79% reduzindo para 18,35%. indicando que as estratégias serão eficazes se aplicadas.
